## 1 προεπεξεργασία  δεδομένων

```python
import json
import pandas as pd #The pandas library
from pandas import Series, DataFrame #Main pandas data structures
import matplotlib.pyplot as plt #main plotting tool for python
import matplotlib as mpl
import numpy as np
import seaborn as sns #A more fancy plotting library
data=[]

with open(r'./meta_Cell_Phones_and_Accessories.json') as f:
    for line in f:
        json_line = json.loads(line)
        
            
        if json_line['category'] and (json_line['category'][-1]== "Car Chargers"or json_line['category'][-1]=="Chargers & Power Adapters")  :
            if not any(d['asin'] == json_line['asin'] for d in data):
                #concat
                description = ' '.join(json_line['description'])
                
                data.append({'description': description, 'category': json_line['category'][-1],'asin' :json_line['asin']})
                #print(data, "21213" )
                
            
           
df1=pd.DataFrame(data)
dfx = df1[['category', 'description']]

print(dfx)
```

## 2 TF-IDF

```python
import sklearn as sklearn
import sklearn.linear_model as linear_model
from sklearn import svm
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
import sklearn.feature_extraction.text as sk_text
import sklearn.metrics as metrics
from sklearn import svm

#Initialize models and performance variables
svm_clf = svm.SVC()
kf = KFold(n_splits=5, shuffle=True)
mlp_clf = MLPClassifier(solver='lbfgs')
 

logReg = linear_model.LogisticRegression()
logreg_confusion_matrix=0
logreg_accuracy=0
logreg_precision=0
logreg_recall=0
logreg_f1=0
svm_confusion_matrix=0
svm_accuracy=0
svm_precision=0
svm_recall=0
svm_f1=0
MLP_accuracy=0
MLP_confusion_matrix=0
MLP_precision=0
MLP_recall=0
MLP_f1=0
for train_index, test_index in kf.split(dfx):
    #Split dfx into training and testing sets
    X_train, X_test = dfx.iloc[train_index], dfx.iloc[test_index]
    y_train, y_test = dfx['category'].iloc[train_index], dfx['category'].iloc[test_index]

    
    vectorizer = sk_text.TfidfVectorizer(max_features=1000, stop_words="english",min_df=4, max_df=0.8)
    
    
    X_train_transformed = vectorizer.fit_transform(X_train['description'])
    X_test_transformed = vectorizer.fit_transform(X_test['description'])
   
    
    #train
    logReg.fit(X_train_transformed, y_train)
    
    y_pred = logReg.predict(X_test_transformed)
    #print("k")
    logreg_accuracy+=metrics.accuracy_score(y_test,y_pred)
    #print("\naccuracy:",metrics.accuracy_score(y_test,y_pred))
    #print("\nConfusion matrix")
    logreg_confusion_matrix+= metrics.confusion_matrix(y_test,y_pred)
    #print(metrics.confusion_matrix(y_test,y_pred))
    #print("\nPrecision Score per class")
    logreg_precision+=metrics.precision_score(y_test,y_pred,average='macro')
    #print(metrics.precision_score(y_test,y_pred,average='macro'))
    #print("\nRecall Score per class")
    logreg_recall+=metrics.recall_score(y_test,y_pred,average='macro')
    #print(metrics.recall_score(y_test,y_pred,average='macro'))
    #print("\nF1-score Score per class")
    logreg_f1+=metrics.f1_score(y_test,y_pred,average='macro')
    #print(metrics.f1_score(y_test,y_pred,average='macro'))
    
    
    
    svm_clf.fit(X_train_transformed,y_train)
    #print("classifier score:",svm_clf.score(X_test_transformed,y_test))
    y_pred = svm_clf.predict(X_test_transformed)
    svm_accuracy+=metrics.accuracy_score(y_test,y_pred)
    #print("\naccuracy:",metrics.accuracy_score(y_test,y_pred))
    #print("\nConfusion matrix")
    svm_confusion_matrix+=metrics.confusion_matrix(y_test,y_pred)
    #print(metrics.confusion_matrix(y_test,y_pred))
    #print("\nPrecision Score per class")
    svm_precision+=metrics.precision_score(y_test,y_pred,average='macro')
    #print(metrics.precision_score(y_test,y_pred,average='macro'))
    #print("\nRecall Score per class")
    svm_recall+=metrics.recall_score(y_test,y_pred,average='macro')
    #print(metrics.recall_score(y_test,y_pred,average='macro'))
    #print("\nF1-score Score per class")
    svm_f1+=metrics.f1_score(y_test,y_pred,average='macro')
    #print(metrics.f1_score(y_test,y_pred,average='macro'))
    
    
    

    
    mlp_clf.fit(X_train_transformed, y_train)
    #print("classifier score:",mlp_clf.score(X_test_transformed,y_test))
    y_pred = mlp_clf.predict(X_test_transformed)
    MLP_accuracy+=metrics.accuracy_score(y_test,y_pred)
    #print("\naccuracy:",metrics.accuracy_score(y_test,y_pred))
    #print("\nConfusion matrix")
    MLP_confusion_matrix+=metrics.confusion_matrix(y_test,y_pred)
    #print(metrics.confusion_matrix(y_test,y_pred))
    #print("\nPrecision Score per class")
    MLP_precision+=metrics.precision_score(y_test,y_pred,average='macro')
    #print(metrics.precision_score(y_test,y_pred,average='macro'))
    #print("\nRecall Score per class")
    MLP_recall+=metrics.recall_score(y_test,y_pred,average='macro')
    #print(metrics.recall_score(y_test,y_pred,average='macro'))
    #print("\nF1-score Score per class")
    MLP_f1+=metrics.f1_score(y_test,y_pred,average='macro')
    #print(metrics.f1_score(y_test,y_pred,average='macro'))
    
logreg_confusion_matrix=logreg_confusion_matrix/5
print("logreg average confusuon matrix :", logreg_confusion_matrix)
logreg_accuracy=logreg_accuracy/5
print("logreg average accuracy :", logreg_accuracy)
logreg_precision=logreg_precision/5
print("logreg average precision :", logreg_precision)
logreg_recall=logreg_recall/5
print("logreg average recall :", logreg_recall)
logreg_f1=logreg_f1/5
print("logreg average f1 :", logreg_f1)
svm_confusion_matrix=svm_confusion_matrix/5
print("svm average confusion matrix :", svm_confusion_matrix)
svm_accuracy=svm_accuracy/5
print("svm average accuracy :", svm_accuracy)
svm_precision=svm_precision/5
print("svm average precision :", svm_precision)
svm_recall=svm_recall/5
print("svm average recall :", svm_recall)
svm_f1=svm_f1/5
print("svm average f1 :", svm_f1)
MLP_confusion_matrix=MLP_confusion_matrix/5

print("MLP average confusion matrix :", MLP_confusion_matrix)
MLP_accuracy=MLP_accuracy/5
print("MLP average accuracy :", MLP_accuracy)
MLP_precision=MLP_precision/5
print("MLP average precision :", MLP_precision)
MLP_recall=MLP_recall/5
print("MLP average recall :", MLP_recall)
MLP_f1=MLP_f1/5
print("MLP average f1 :", MLP_f1)
```

## 3 DOC2VEC

```python
import gensim 
import gensim.models
from gensim.models.doc2vec import Doc2Vec,TaggedDocument
from gensim import utils

#Initialize models and performance variables
logreg_confusion_matrix=0
logreg_accuracy=0
logreg_precision=0
logreg_recall=0
logreg_f1=0
svm_confusion_matrix=0
svm_accuracy=0
svm_precision=0
svm_recall=0
svm_f1=0
MLP_accuracy=0
MLP_confusion_matrix=0
MLP_precision=0
MLP_recall=0
MLP_f1=0



for train_index, test_index in kf.split(dfx):
    #Split dfx into training and testing sets 
    X_train, X_test = dfx.iloc[train_index], dfx.iloc[test_index]
    y_train, y_test = dfx['category'].iloc[train_index], dfx['category'].iloc[test_index]

    #Preprocessing training data
    train_gsim = [utils.simple_preprocess(x) for x in X_train['description']]
    train_data_labels = [(x,y) for (x,y) in zip(train_gsim, y_train) if len(x) > 0]
    X_news_train_gsim = [x for (x,y) in train_data_labels]
    y_news_train_gsim = [y for (x,y) in train_data_labels]

    #Preprocessing test data 
    test_gsim = [utils.simple_preprocess(x) for x in X_test['description']]
    test_data_labels = [(x,y) for (x,y) in zip(test_gsim, y_test) if len(x) > 0]
    X_news_test_gsim = [x for (x,y) in test_data_labels]
    y_news_test_gsim = [y for (x,y) in test_data_labels]

    train_corpus = [gensim.models.doc2vec.TaggedDocument(X_news_train_gsim[i], [i]) for i in range(len(X_news_train_gsim))]

    d2v_model = gensim.models.doc2vec.Doc2Vec(vector_size=50, min_count=2, epochs=40)
    d2v_model.build_vocab(train_corpus)
    d2v_model.train(train_corpus, total_examples=d2v_model.corpus_count, epochs=d2v_model.epochs)

    X_news_train_d2v = [d2v_model.infer_vector(x) for x in X_news_train_gsim]
    X_news_test_d2v = [d2v_model.infer_vector(x) for x in X_news_test_gsim]
    
    
    logReg.fit(X_news_train_d2v, np.array(y_news_train_gsim))
    logReg.score(X_news_test_d2v, y_news_test_gsim)
    
    
    
    
    
    logReg.fit(X_news_train_d2v, y_news_train_gsim)
    
    y_pred = logReg.predict(X_news_test_d2v)
    #print("k")
    logreg_accuracy+=metrics.accuracy_score(y_news_test_gsim,y_pred)
    #print("\naccuracy:",metrics.accuracy_score(y_test,y_pred))
    #print("\nConfusion matrix")
    logreg_confusion_matrix+= metrics.confusion_matrix(y_news_test_gsim,y_pred)
    #print(metrics.confusion_matrix(y_test,y_pred))
    #print("\nPrecision Score per class")
    logreg_precision+=metrics.precision_score(y_news_test_gsim,y_pred,average='macro')
    #print(metrics.precision_score(y_test,y_pred,average='macro'))
    #print("\nRecall Score per class")
    logreg_recall+=metrics.recall_score(y_news_test_gsim,y_pred,average='macro')
    #print(metrics.recall_score(y_test,y_pred,average='macro'))
    #print("\nF1-score Score per class")
    logreg_f1+=metrics.f1_score(y_news_test_gsim,y_pred,average='macro')
    #print(metrics.f1_score(y_test,y_pred,average='macro'))
    
    
    
    svm_clf.fit(X_news_train_d2v, y_news_train_gsim)
    #print("classifier score:",svm_clf.score(X_test_transformed,y_test))
    y_pred = svm_clf.predict(X_news_test_d2v)
    svm_accuracy+=metrics.accuracy_score(y_news_test_gsim,y_pred)
    #print("\naccuracy:",metrics.accuracy_score(y_test,y_pred))
    #print("\nConfusion matrix")
    svm_confusion_matrix+=metrics.confusion_matrix(y_news_test_gsim,y_pred)
    #print(metrics.confusion_matrix(y_test,y_pred))
    #print("\nPrecision Score per class")
    svm_precision+=metrics.precision_score(y_news_test_gsim,y_pred,average='macro')
    #print(metrics.precision_score(y_test,y_pred,average='macro'))
    #print("\nRecall Score per class")
    svm_recall+=metrics.recall_score(y_news_test_gsim,y_pred,average='macro')
    #print(metrics.recall_score(y_test,y_pred,average='macro'))
    #print("\nF1-score Score per class")
    svm_f1+=metrics.f1_score(y_news_test_gsim,y_pred,average='macro')
    #print(metrics.f1_score(y_test,y_pred,average='macro'))
    
    
    

    
    mlp_clf.fit(X_news_train_d2v, y_news_train_gsim)
    #print("classifier score:",mlp_clf.score(X_test_transformed,y_test))
    y_pred = mlp_clf.predict(X_news_test_d2v)
    MLP_accuracy+=metrics.accuracy_score(y_news_test_gsim,y_pred)
    #print("\naccuracy:",metrics.accuracy_score(y_test,y_pred))
    #print("\nConfusion matrix")
    MLP_confusion_matrix+=metrics.confusion_matrix(y_news_test_gsim,y_pred)
    #print(metrics.confusion_matrix(y_test,y_pred))
    #print("\nPrecision Score per class")
    MLP_precision+=metrics.precision_score(y_news_test_gsim,y_pred,average='macro')
    #print(metrics.precision_score(y_test,y_pred,average='macro'))
    #print("\nRecall Score per class")
    MLP_recall+=metrics.recall_score(y_news_test_gsim,y_pred,average='macro')
    #print(metrics.recall_score(y_test,y_pred,average='macro'))
    #print("\nF1-score Score per class")
    MLP_f1+=metrics.f1_score(y_news_test_gsim,y_pred,average='macro')
    #print(metrics.f1_score(y_test,y_pred,average='macro'))
    
logreg_confusion_matrix=logreg_confusion_matrix/5
print("logreg average confusuon matrix :", logreg_confusion_matrix)
logreg_accuracy=logreg_accuracy/5
print("logreg average accuracy :", logreg_accuracy)
logreg_precision=logreg_precision/5
print("logreg average precision :", logreg_precision)
logreg_recall=logreg_recall/5
print("logreg average recall :", logreg_recall)
logreg_f1=logreg_f1/5
print("logreg average f1 :", logreg_f1)
svm_confusion_matrix=svm_confusion_matrix/5
print("svm average confusion matrix :", svm_confusion_matrix)
svm_accuracy=svm_accuracy/5
print("svm average accuracy :", svm_accuracy)
svm_precision=svm_precision/5
print("svm average precision :", svm_precision)
svm_recall=svm_recall/5
print("svm average recall :", svm_recall)
svm_f1=svm_f1/5
print("svm average f1 :", svm_f1)
MLP_confusion_matrix=MLP_confusion_matrix/5

print("MLP average confusion matrix :", MLP_confusion_matrix)
MLP_accuracy=MLP_accuracy/5
print("MLP average accuracy :", MLP_accuracy)
MLP_precision=MLP_precision/5
print("MLP average precision :", MLP_precision)
MLP_recall=MLP_recall/5
print("MLP average recall :", MLP_recall)
MLP_f1=MLP_f1/5
print("MLP average f1 :", MLP_f1)
    
    
```
